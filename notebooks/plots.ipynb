# Tiempos

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sample data for PaddleOCR time comparison by document type
data = {
    'Document': ['INE', 'CFE', 'TELMEX', 'Promedio'],
    'Baseline (s)': [2.06, 2.06, 2.10, 2.04],  # Converted from ms to seconds
    'New Time (s)': [1.01, 1.10, 1.15, 1.3]  # Converted from ms to seconds
}

# Create DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame for easier plotting with seaborn
df_melted = pd.melt(df, id_vars=['Document'], 
                    value_vars=['Baseline (s)', 'New Time (s)'],
                    var_name='Method', value_name='Time (s)')

# Create bar plot
plt.figure(figsize=(12, 7))
ax = sns.barplot(x='Document', y='Time (s)', hue='Method', data=df_melted, palette='viridis')

# Add title and labels
plt.title('PaddleOCR Processing Time Comparison by Document Type - 4 workers', fontsize=16)
plt.xlabel('Document Type', fontsize=14)
plt.ylabel('Time (seconds)', fontsize=14)

# Add value labels on top of bars
for i, p in enumerate(ax.patches):
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 0.005,
            f'{height:.3f}', ha='center', fontsize=11)

# Customize grid and style
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.despine(left=False, bottom=False)
plt.legend(title='Method', fontsize=12, title_fontsize=13)

plt.tight_layout()
plt.show()
```
